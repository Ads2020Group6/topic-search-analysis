## 2 - Relative Topic Interest

We compare interest over different *Topics*.
- Gender Pay Gap
- Gender Equality
- Data Science

```python
topics_interest = ps.read_csv("TopicInterest.csv")
print(topics_interest.shape)
topics_interest = topics_interest.set_index("Month")
topics_interest.head()
```

```python
topics_interest['Data science'] = topics_interest['Data science'].replace('<1','1')
topics_interest['Data science'] = topics_interest['Data science'].apply(lambda x: int(x))
# topics_interest['Data science']['2006-12']
# topics_interest['Data science']['2006-01']
# topics_interest['Data science']['2010-12']
topics_interest.dtypes
```

```python
topics_interest.plot(figsize=(13, 6))
```

## 3 - Interest by region

Locations where Gender Pay Gap is more popular

```python
topic_by_region = ps.read_csv("TopicInterestMap.csv")
print(topic_by_region.shape)
topic_by_region = topic_by_region.set_index("Country")
topic_by_region = topic_by_region.sort_values("Gender pay gap", ascending = False)
topic_by_region["Gender pay gap"] = topic_by_region["Gender pay gap"].fillna(0)
print(topic_by_region.head())
topic_by_region.tail()
```
